```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
answers = pd.read_csv('Answer.csv', encoding="utf-16")
answers = answers.drop(0) # remove the meaningless row(the second line)
answers.head(10)
```

```python
len(answers['ClosedDate'].dropna())
```

```python
questions = pd.read_csv('Questions.txt', encoding="utf-16")
questions.head(3)
```

```python
Q_A = pd.read_csv('Q_A.csv', encoding="utf-16")
Q_A.head(3)
```

```python
Q_A[Q_A['accepted']==1]
```

```python
users = pd.read_csv('U.csv', encoding="utf-16")
users.head(3)
```

```python
user_badge = pd.read_csv('user_badge.txt', encoding="utf-16")
user_badge.head(10)
```

```python
tags = pd.read_csv('tags.txt', sep='\t')
tags.head(10)
```

```python
len(questions)
```

```python
len(answers)
```

```python
merged_A_Q = pd.merge(answers, Q_A, left_on='Id', right_on='AId', how='inner')
merged_A_Q = pd.merge(merged_A_Q, questions, left_on='QId', right_on='Id', how='inner',suffixes=('_A', '_Q'))
merged_A_Q.head()
```

```python
len(merged_A_Q)
```

```python
merged_A_Q_U = pd.merge(merged_A_Q, users,  left_on='OwnerUserId_A', right_on='Id', how='inner')
merged_A_Q_U.head()
```

```python
len(merged_A_Q_U)
```

```python
merged_A_Q_U2 = pd.merge(merged_A_Q, users,  left_on='OwnerUserId_A', right_on='Id', how='left')
merged_A_Q_U2.head()
```

```python
len(merged_A_Q_U2)
```

```python
merged_A_Q_U2[merged_A_Q_U2['Id'].isnull()]
```

```python
badge_count = user_badge.groupby('UserId').size().reset_index(name='badgeCount')
badge_count
```

```python
merged_A_Q_U_B = pd.merge(merged_A_Q_U, badge_count, left_on='Id', right_on='UserId', how='inner')
merged_A_Q_U_B.head()
```

```python
len(merged_A_Q_U_B)
```

```python
merged_A_Q_U_B['ClosedDate_Q']
```

```python
merged_A_Q_U_B['Closed_Q'] = merged_A_Q_U_B['ClosedDate_Q'].notnull().astype(int)
```

```python
merged_A_Q_U_B['Closed_Q']
```

```python
merged_A_Q_U_B.columns
```

```python
merged_A_Q_U_B['CreationDate_A'] = pd.to_datetime(merged_A_Q_U_B['CreationDate_A'])
merged_A_Q_U_B['CreationDate_Q'] = pd.to_datetime(merged_A_Q_U_B['CreationDate_Q'])
merged_A_Q_U_B['timeDiffrence(Hours)'] = (merged_A_Q_U_B['CreationDate_A'] - merged_A_Q_U_B['CreationDate_Q']).dt.total_seconds() / 60 / 60 #based on hours
merged_A_Q_U_B['timeDiffrence(Hours)']
```

```python
merged_A_Q_U_B[merged_A_Q_U_B['timeDiffrence(Hours)']==0]
```

```python
merged_A_Q_U_B[merged_A_Q_U_B['timeDiffrence(Hours)']<0]
```

```python
len(merged_A_Q_U_B)
```

```python
merged_A_Q_U_B2 = merged_A_Q_U_B[merged_A_Q_U_B['timeDiffrence(Hours)']>0]
```

```python
len(merged_A_Q_U_B2)
```

```python
merged_A_Q_U_B2.columns
```

```python
df_answers = merged_A_Q_U_B2.groupby('UserId').size().reset_index(name='#_answers')
df_answers
```

```python
df_accepted = merged_A_Q_U_B2[merged_A_Q_U_B2['accepted']==1].groupby('UserId').size().reset_index(name='#_accepted')
df_accepted
```

```python
merged_df = pd.merge(df_answers, df_accepted,  left_on='UserId', right_on='UserId', how='left')
merged_df
```

```python
merged_df['#_accepted'] = merged_df['#_accepted'].fillna(0).astype(int)
```

```python
merged_df[merged_df['#_answers'].isnull()]
```

```python
merged_df[merged_df['#_answers']==0]
```

```python
merged_df
```

```python
merged_df['accepted_ratio'] = (merged_df['#_accepted']/merged_df['#_answers'])
merged_df
```

```python
merged_A_Q_U_B2_R = pd.merge(merged_A_Q_U_B2, merged_df,  left_on='OwnerUserId_A', right_on='UserId', how='left')
merged_A_Q_U_B2_R
```
